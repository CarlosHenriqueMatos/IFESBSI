## Exemplo com uma base de dados do Kaggle
Base de dados: https://www.kaggle.com/datasets/shazadudwadia/supermarket?resource=download

Artigo explicando: https://medium.com/@mervetorkan/association-rules-with-python-9158974e761a


```python
#para  importar os arquivos que estão na minha pasta local do google drive eu utilizo o seguinte codigo
baseDeDados = pd.read_csv('/content/drive/MyDrive/Ifes/2022_2/Inteligência de Negócios - Kelly/Entrega 2/GroceryStoreDataSet.csv', names = ['Produtos'], sep = ',')
baseDeDados.head()
```

```python
#Dimensões do codigo
baseDeDados.shape
```

```python
dadosComSplit = list(baseDeDados["Produtos"].apply(lambda x:x.split(",") ))
dadosComSplit
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_dados = a.fit(dadosComSplit).transform(dadosComSplit)
baseDeDados = pd.DataFrame(a_dados,columns=a.columns_)
baseDeDados = baseDeDados.replace(False,0)
baseDeDados = baseDeDados.replace(True,1)
baseDeDados
```

```python
baseDeDados = apriori(baseDeDados, min_support = 0.2, use_colnames = True)
baseDeDados
```

```python
df_ar = association_rules(baseDeDados, metric = "confidence", min_threshold = 0.6)
df_ar
```
